# Diagnosing Pneumonia from Chest Xray Images and Convolutional Neural Networks 

## Importing All Necessary Packages:

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import imageio
import cv2

import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from tensorflow.keras import layers
from keras.layers import Dense, Conv2D, MaxPool2D, Dropout, Flatten, BatchNormalization
from keras.callbacks import ReduceLROnPlateau, EarlyStopping
```

## Asigning Data Paths to variables:

```python
train_dir = './data/train'
test_dir = './data/test'
val_dir = './data/val'

train_norm_dir = train_dir + '/NORMAL/'
train_pneumo_dir = train_dir + '/PNEUMONIA/'

test_norm_dir = test_dir + '/NORMAL/'
test_pneumo_dir = test_dir + '/PNEUMONIA/'

val_norm_dir = val_dir + '/NORMAL/'
val_pneumo_dir = val_dir + '/PNEUMONIA/'
```

## Importing Data: 

```python
#Creating function to import images from file directories while keeping their respective labels

def import_imgs(directory):
    labels = ['NORMAL', 'PNEUMONIA']
    img_sz = 256
    data_lst = []
    for l in labels:
        path = os.path.join(directory, l)
        class_num = labels.index(l)
        for img in os.listdir(path):    
            img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
            resized_arr = cv2.resize(img_arr, (img_sz, img_sz)) 
            data_lst.append([resized_arr, class_num])
         
    return np.array(data_lst)
```

```python
# using above defined function to create arrays of image/label for the train, test and validation sets

train = import_imgs(train_dir)
test = import_imgs(test_dir)
val = import_imgs(val_dir)
```

```python
#Splitting up the feature image data (x) from the target label data (y)

x_train = []
y_train = []
for feat, label in train:
    x_train.append(feat)
    y_train.append(label)
    
x_test = []
y_test = []

for feat, label in test:
    x_test.append(feat)
    y_test.append(label)
    
x_val = []
y_val = []

for feat, label in val:
    x_val.append(feat)
    y_val.append(label)
```

# Building the CNN Model:

### Determining Data Class Weights:

```python
# due to imbalance in data we must first determine weights for each catergory

weight_0 = (train_pneumo_num + train_norm_num) / (2 * train_norm_num)
weight_1 = (train_pneumo_num + train_norm_num) / (2 * train_pneumo_num)

class_weights = {0: weight_0, 1: weight_1}

print(f'Class 0 has weight of {weight_0:.2f} and class 1 has weight of {weight_1:.2f}')
```

## Building the Model:

```python
model = Sequential()
model.add(Conv2D(32 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu' , input_shape = (256,256,1)))
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.2))
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(128 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.2))
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(256 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.2))
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Flatten())
model.add(Dense(units = 128 , activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 1 , activation = 'sigmoid'))
model.compile(optimizer = "adam" , loss = 'binary_crossentropy' , metrics = ['accuracy'])
model.summary()
```

```python
model = Sequential()
model.add(Conv2D(32 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu' , input_shape = (256,256,1)))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.1))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(128 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Conv2D(256 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
model.add(Flatten())
model.add(Dense(units = 128 , activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 1 , activation = 'sigmoid'))
model.compile(optimizer = "adam" , loss = 'binary_crossentropy' , metrics = ['accuracy'])
model.summary()
```

## Fitting the Model:

```python
#Creating a callback for the model fit so that learning rate is reduced after a number of epochs reached with no further increase in validation accuracy
#Another to stop fitting early if no decrease in val_loss seen

learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy', patience = 2, verbose=1,factor=0.3, min_lr=0.000001)

early_stopping = EarlyStopping(monitor="val_loss", patience = 5)
```

```python
history = model.fit(train_gen, epochs = 10 , validation_data = val_gen, callbacks = [learning_rate_reduction, early_stopping], class_weight=class_weights)

```

```python
model.evaluate(test_gen)
```

```python

train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

plt.plot(train_acc , 'bo-', label = 'Training Accuracy')
plt.plot(val_acc , 'ro-', label = 'Validation Accuracy')
plt.title('Training & Validation Accuracy')
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Accuracy");
```

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(train_loss, 'bo-', label = 'Training Loss')
plt.plot(val_loss, 'ro-', label = 'Validation Loss')
plt.title('Training & Validation Loss')
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("loss");
```
